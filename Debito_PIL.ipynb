### Time series dataset relating to Public Debt as a percentage of GDP for the following European countries: 
* Italy, France, Spain and Germany.

#### Data source: (https://ec.europa.eu/economy_finance/ameco/user/serie/ResultSerie.cfm) 


```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 
```

```python
df = pd.read_excel('DEBITOPIL_4.xlsx', index_col=0)
```

```python
# Debt as (%ofPIL) 
# Year available for Italy: 1960 to 2020
# Year available for Germany, Spain, France: 1995 to 2021

df.info()
```

```python
df.head()
```

```python
# Missing value for Germany, Spain and France in the years (1960-1995)

sns.heatmap(pd.isnull(df), cmap='coolwarm')
```

```python
#Non-missing dataset

country = df[df['Germany']>0]
```

### Timeseries plot

```python
# Countries togheter (1960-2021)

sns.set()

df.plot(figsize=(14,6), title='Public Debt as % of PIL',cmap='gist_rainbow')
plt.ylabel('Public Debt/PIL ratio')
```

```python
# Italy (1960-2021)

sns.set()

df['Italy'].plot(figsize=(14,6), title='Public Debt as % of PIL',legend=True,cmap='gist_rainbow')
plt.ylabel('Public Debt/PIL ratio')
```

```python
# All countries (1995-2021)

sns.set()

country.plot(figsize=(14,6), title='Public Debt as % of PIL',legend=True,cmap='gist_rainbow')
plt.ylabel('Public Debt/PIL ratio')
```
