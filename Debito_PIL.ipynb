```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot

%matplotlib inline 
```

```python
df = pd.read_excel('DEBITOPIL_4.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
type(df)
```

### seaborn graphic

```python
plt.figure(figsize=(12,4))

sns.set(style="whitegrid" , palette='bright', font_scale=1)

sns.lineplot(x='Year', y='Italy',data=df, palette="tab10", linewidth=2.5 )

plt.autoscale(enable=True, axis='both')

plt.legend(title='', loc='upper left', labels=['Italia'])
plt.title('Debito pubblico in percentuale del PIL')
plt.xlabel('Anno')
plt.ylabel('Rapporto Debito/Pil')

plt.savefig("images/output1.png")
```
